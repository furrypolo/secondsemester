```python
import numpy as np

def rumus(x, n):
    atas = (-1) ** n
    bawah = np.math.factorial(2 * n)
    samping = x ** (2 * n)
    hasil = atas / bawah * samping
    return hasil
```

```python
def hitung_cos(x):
    total = 0

    list_nilai = []
    for i in range(10):
        total += rumus(x, i)
        list_nilai.append(total)
    
    return list_nilai
```

```python
# Cos(0)
nilai_y = hitung_cos(0)

import matplotlib.pyplot as plt

plt.plot(nilai_y, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("cos(0)")
```

```python
# Cos(pi/4)
nilai_y = hitung_cos(np.pi / 4)

import matplotlib.pyplot as plt

plt.plot(nilai_y, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("cos(pi/4)")
```

```python
# Cos(pi/3)
nilai_y = hitung_cos(np.pi / 3)

import matplotlib.pyplot as plt

plt.plot(nilai_y, marker="o", color="g", linestyle=":")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("cos(pi/3)")
```
