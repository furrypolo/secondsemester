```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (0.5*x*(np.power(np.e, np.power(x, 0.5)))) - 2

def bissection(xi, xu, err, errs):
    list_error=[]
    iter = 0
    while(err>errs):
        iter+=1;
        print("Iterasi ke-%d" %(iter))
        print("xi: %.4f, xu : %.4f" %(xi, xu))
        print("f(xi): %.4f, f(xu): %.4f" %(f(xi), f(xu)))
        xr = (xi+xu)/2
        print()

        if(iter>1):
            err = abs((xr-xr_old)/xr)*100
            print("error : %.4f, " %err)
            list_error.append(err)

        if(f(xi) * f(xr)<0):
            xu = xr
        elif(f(xi) * f(xr)>0):
            xi = xr
        else:
            break

        xr_old = xr
    
    plt.plot(range(2, iter+1), list_error, marker="o")
    plt.ylabel("error")
    plt.xlabel("loop")
    plt.grid()
    plt.axis([0,6,0,25])
    plt.show()
```

```python
bissection(1,2,100,4)
```
