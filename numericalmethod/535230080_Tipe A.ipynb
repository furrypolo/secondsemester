```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#No. 1
def f(x):
    return (0.5*np.power(np.e, np.sqrt(((1/3)*x))))-1
```

```python
def error(last, now):
    fin = abs((now-last)/now)*100
    return fin
```

```python
def regulafasi(xi, xu, es):
    err = 100
    bef = 0
    iterasi = 0
    xrr = []
    e = []
    while(err>es):
        iterasi += 1
        fxi = f(xi)
        fxu = f(xu)
        xr = xu - ((fxu * (xi-xu))/(fxi-fxu))
        fxr = f(xr)
        if(bef!=0):
            err = error(bef, xr)
        else:
            err = '-'
        bef = xr
        xrr.append(xr)
        print('iterasi ke-',iterasi)
        print('xi : ', xi)
        print('xu : ', xu)
        print('f(xi) : ', fxi)
        print('f(xu) : ', fxu)
        print('xr : ', xr)
        print('f(xr) : ', fxr)
        print('error : ', err)
        print()
        if(err == '-'):
            err = 100
        else :
            e.append(err)
        

        if(fxr>0):
            xu = xr
        else:
            xi = xr
    
    plt.title("Akar Persamaan dg Regulafasi")
    plt.subplot(2,1,1)
    plt.plot(range(1, iterasi+1), xrr, marker='o', linestyle='None')
    plt.ylabel('Hasil Regulafasi')
    plt.axis()
    plt.grid()
    plt.subplot(2,1,2)
    plt.plot(range(2, iterasi+1), e, marker='o', linestyle='None')
    plt.ylabel("Error")
    plt.grid()
    plt.show()

```

```python
regulafasi(1, 50, 0.05)
```

```python
#No. 2
def lagrange(x, y, orde, n):
    minimum = 0
    maximum = 0
    for t in range (len(x)):
        if((x[t]<n) & (x[t+1]>n)):
            min=t
            max=t+1
            break

    xt=[]
    yt=[]
    minimum = 0
    maximum = 0
    if(min==0):
        if((n-x[min])<(x[max]-n)):
            minimum = min
            maximum = minimum+orde
        else:
            minimum = max
            maximum = max + orde
    else:
        minimum = min
        maximum = max
        for t in range (orde+1):
            if((x[minimum]-x[minimum-1])<(x[maximum+1]-x[maximum])):

                minimum = minimum-1
                maximum = maximum
            else:
                minimum = minimum
                maximum = maximum+1
    

    
    for t in range (minimum,maximum+1):
        xt.append(x[t])
        yt.append(y[t])

    
    L=1
    f=0

    for t in range (orde+1):
        L=1
        for m in range (orde+1):
            if(m!=t):

                L *= ((n-xt[m])/(xt[t]-xt[m]))

        f += L*yt[t]
    
    print("Pencarian nilai titik ", n, " dengan Lagrange orde ", orde)
    print("Menggunakan nilai-nilai: ")
    for t in range (orde+1):
        print("x",t, " = ", xt[t])
    print("Hasil = ", f)
    print(f)
```
