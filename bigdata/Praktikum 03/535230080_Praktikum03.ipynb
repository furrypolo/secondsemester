```python
#Mencari film dengan durasi terpanjang dari semua data
longest = moviesClean[moviesClean['Duration']==moviesClean['Duration'].max()]

#Mencari film dengan durasi terpendek dari semua data
shortest = moviesClean[moviesClean['Duration']==moviesClean['Duration'].min()]

#Menggabungkan kedua hasil data dan menampilkannya
pd.concat([longest, shortest])
```

```python
#Mengeluarkan baris data yang memenuhi ketentuan Score>8.3 dan Duration<120
moviesClean[(moviesClean['Score']>8.3) & (moviesClean['Duration']<120)]
```

```python
#Solution2 dari mengeluarkan baris data yang memenuhi ketentuan Score>8.3 dan Duration<120
moviesClean.query('Score>8.3 and Duration<120')
```

```python
#Solution 3 dari mengeluarkan baris data yang memenuhi ketentuan Score>8.3 dan Duration<120
moviesClean.query('Score>8.3 & Duration<120')
```

```python
#Mengeluarkan baris data yang memenuhi ketentuan Director berisi nama Ridley Scott
moviesClean[moviesClean['Director']=="Ridley Scott"]
```

```python
#Solution2 dari mengeluarkan baris data yang memenuhi ketentuan Director berisi nama Ridley Scott
dir="Ridley Scott"
moviesClean.query('Director==@dir')
```

```python
#Solution 3 dari mengeluarkan baris data yang memenuhi ketentuan Director berisi nama Ridley Scott
moviesClean.query('Director=="Ridley Scott"')
```

```python
#Mengeluarkan baris data yang memenuhi ketentuan dimana Actor 1 atau Actor 2 atau Actor 3 berisi nama "Hugh Jackman"
moviesClean[(moviesClean['Actor 2'] == "Hugh Jackman") | (moviesClean['Actor 3'] == "Hugh Jackman") | (moviesClean['Actor 1'] == "Hugh Jackman")]
```

```python
#Mengumpulkan film berdasarkan kelompok value kolom 'Colour' dan menampilkan jumlah dari anggota baris data
moviesClean.groupby(['Colour']).sum()

#atau dengan cara=
moviesClean.groupby('Colour').sum()
```

```python
#Menampilkan baris yang memenuhi ketentuan dimana:
moviesClean[((moviesClean['Year']>=2005) & (moviesClean['Year'] <= 2010)) #Year diantara 2005 sampai 2010
& (moviesClean['Score']>8.0)  #Score diatas 8.0
& (moviesClean['Gross Income'] < moviesClean['Budget'])] #Dan Gross Income kurang dari Budget yang ditentukan
```

```python
#Menampilkan semua baris data yang memenuhi ketentuan:
moviesClean[
    (moviesClean['Year'] >= 2010) & #Year diatas atau sama dengan 2010
    (moviesClean['Score'] > 8.0) & #Memiliki Score lebih besar dari 8.0
    (moviesClean['Gross Income'] < moviesClean['Gross Income'].quantile(0.25)) #Dan dibawah persentil ke 25 dalam pendapatan
]
```

```python
#Mengeluarkan jumlah Votes, Budget, dan Gross Income setiap tahunnya
moviesClean.groupby(['Year'])['Votes', 'Budget', 'Gross Income'].sum()
```

```python
#Menampilkan 10 sutradara teratas dengan rata rata Score tertinggi
moviesClean.groupby('Director')['Score'].mean().sort_values(ascending=False)[0:10]
#ascending=False agar dia muncul dari terbesar ke terkecil
#[0:10] agar dia hanya memunculkan 10 baris data
```
